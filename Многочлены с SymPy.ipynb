# 2. Многочлены с SymPy

Многочленом называется функция вида
$$
P_n(z) = a_n z^n + a_{n-1}z^{n-1} + ... + a_1 z + a_0,
$$
где $a_0$, $a_1$, $...$, $a_n$ $-$ фиксированные комплексные числа, коэффициенты многочлена, $n\ge 0$ $-$ степень многочлена.

Число $\alpha$ называется корнем многочлена, если $P_n(\alpha)=0$.

Число $\alpha$ называется корнем кратности $k\ge 1$ многочлена $P_n$, если $P_n(z)$ делится на $(z-\alpha)^k$ без остатка,
$$
P_n(z)=(z-\alpha )^kq_{n-k}(z),
$$
а частное $q_{n-k}(z)$ не делится на $(z-\alpha )$, т.е. $\alpha$ не является корнем многочлена $q_{n-k}(z)$. Если кратность корня равна единице, то корень называют простым.

Всякий многочлен степени $n\ge 1$ имеет ровно $n$ корней с учетом их кратности, и справедливо следующее разложение многочлена на множители:
$$
	P_n(z)=(z-\alpha_1)^{k_1}(z-\alpha_2)^{k_2}\cdots(z-\alpha_m)^{k_m},
$$
где
$$k_1+k_2+\cdots+k_m=n.$$

Для того чтобы найти все корни  многочлена степени $n$ (такого, что явной формулы нет) без использования компьютера, нужно сначала догадаться, какое число может быть корнем. Затем надо убедиться, что данное число $\alpha$ действительно является корнем, и разделить многочлен на ${z-\alpha}$. Эти действия следует повторить для частного, и так до тех пор, пока не будут найдены все $n$ корней с учетом их кратности.


**Упражнение 1.** Получить список чисел, подозрительных на то, что они могут быть корнями многочлена
$$
	P_5(z)=z^5-5z^4+7z^3-2z^2+4z-8.
$$

**Решение.** Согласно теореме Вьета, для корней
$\alpha_1$, $\alpha_2$, ..., $\alpha_n$ приведенного многочлена (все коэффициенты исходного многочлена делятся на его старший коэффициент $a_n$),
$$
\widetilde{P}_n(z) =  z^n + \frac {a_{n-1}}{a_n}z^{n-1} + ... + \frac{a_1}{a_n} z + \frac{a_0}{a_n},
$$
справедливо равенство
$$
\frac{a_0}{a_n}=(-1)^n\alpha_1\alpha_2\cdots\alpha_n.
$$
Поэтому начать поиск его корней
можно  среди  чисел вида $\displaystyle\frac{s}{t}$,
где $s$ $-$ делители $a_0$, а $t$ $-$ делители $a_n$.

Выпишем  делители свободного члена $a_0$ многочлена $P_5(z)$: $\pm 1, \pm 2, \pm 4, \pm 8$.
Делители его старшего коэффициента $a_n$ равны $\pm 1$. Поэтому числами, подозрительными на то, что они могут быть корнями этого многочлена, являются числа $\pm 1, \pm 2, \pm 4, \pm 8$.

**Упражнение 4.** Разложить многочлен $P_5(z)$ на множители.

**Решение.** Мы только что установили, что этот многочлен  можно представить в
виде
$$
P_5(z)=(z-2)^3(z^2+z+1).
$$
Для того, чтобы найти оставшиеся два корня многочлена $P_5(z)$, необходимо решить квадратное уравнение
$$
z^2+z+1=0.
$$
Решения этого уравнения --- комплексные числа
$$
\alpha_{1}=-\frac{1}{2} + i\frac{\sqrt{3}}{2}, \quad \alpha_{2}=-\frac{1}{2}- i\frac{\sqrt{3}}{2}.
$$
Действительно, мы знаем, что числа
\begin{equation*} \alpha_1=p+i\sqrt{q-p^2}, \quad
	\alpha_2=p-i\sqrt{q-p^2}.
\end{equation*}
являются корнями квадратного уравнения, записанного в виде
\begin{equation*}
	z^2-2pz+q=0.
\end{equation*}
Занумеруем все корни многочлена $P_5(z)$ с учетом их кратности:
$$
\alpha_{1,2} = -\frac{1}{2} \pm i\frac{\sqrt{3}}{2}, \quad \alpha_{3,4,5} = 2.
$$
Многочлен $P_5(z)$ можно представить теперь в виде
$$P_5(z) = (z - 2)^3\left(z +
\frac{1}{2} + i\frac{\sqrt{3}}{2}\right)\left(z + \frac{1}{2} - i\frac{\sqrt{3}}{2}\right).$$

**Упражнение 5.** Разложить многочлен $P_5(z)$ на неприводимые вещественные множители.

**Решение**. У многочлена
с вещественными коэффициентами комплексные корни всегда представляют собой пары комплексно сопряженных
чисел, $\alpha$ и $\overline{\alpha}$,  для которых справедливо равенство
$$
(z-\alpha)(z-\overline{\alpha})=z^2+(-2\,\mathrm{Re}\,\alpha)z+|\alpha|^2.
$$
В рассматриваемом примере $\alpha_2=\overline{\alpha}_1$, кроме того,
$$
(z-\alpha_1)(z-\overline{\alpha}_1)=z^2+z+1,
$$
т.е. при вещественных $z$ полином $P_5(z)$ допускает представление в виде произведения трех линейных и
одного квадратичного вещественных сомножителей, а именно,  $P_5(z) = (z - 2)^3(z^2+z+1).$
